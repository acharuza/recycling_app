```python
from PIL import Image
from matplotlib import pyplot as plt
import torchvision

img_in = Image.open("..\\data\\garbage_classification\\Garbage classification\\Garbage classification\\cardboard\\cardboard142.jpg")

plt.imshow(img_in)
```

```python
def unnormalize(tensor, mean, std):
    for t, m, s in zip(tensor, mean, std):
        t.mul_(s).add_(m)
    return tensor

plt.imshow(torchvision.transforms.functional.to_pil_image(unnormalize(img_out, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), mode=None))
```
